```python
from sklearn.datasets import load_iris
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
print(iris.feature_names)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(data[:,0], bw = 0.5)
plt.show()
```

```python
clf = GaussianNB()
clf.fit(data,target)
result = clf.predict(data)
print(result)
```

```python
acc = accuracy_score(target, result, normalize = True) * float(100)
print(str(acc)+' %')
```
