```python
import pandas as pd
```

```python
messages = pd.read_csv('smsspamcollection.csv')
```

```python
messages
```

```python
# data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    words = review.split()
    cleaned_words = [stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    cleaned_review = ' '.join(cleaned_words)
    corpus.append(cleaned_review)
```

```python
corpus
```

```python
# create BOW
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 2500, binary = True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
cv.vocabulary_
```

```python
# N-gram

cv2 = CountVectorizer(max_features = 2500, binary = True, ngram_range=(1,2))
X = cv2.fit_transform(corpus).toarray()
```
